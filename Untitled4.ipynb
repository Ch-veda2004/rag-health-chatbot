```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import SentenceTransformerEmbeddings
from langchain_groq import ChatGroq
from langchain_core.documents import Document

medical_data = [
    "Diabetes symptoms include frequent urination, excessive thirst, and fatigue. Treatment includes insulin therapy.",
    "Hypertension is high blood pressure above 130/80 mmHg. Managed with diet, exercise, and ACE inhibitors.",
    "Asthma causes airway inflammation. Symptoms include wheezing and shortness of breath. Inhalers provide relief.",
    "Depression causes persistent sadness and fatigue. Treatment includes therapy and antidepressants.",
    "COVID-19 symptoms include fever, cough, and loss of taste. Vaccines reduce severe illness risk.",
    "Anemia is lack of healthy red blood cells. Iron deficiency is most common. Causes fatigue and pale skin.",
    "Migraine causes severe headaches, nausea, and light sensitivity. Triptans are commonly prescribed.",
    "Arthritis causes joint inflammation and stiffness. Anti-inflammatory medications help manage symptoms.",
    "Thyroid disorders cause weight changes, fatigue, and mood swings.",
    "Malaria is caused by Plasmodium parasites via mosquitoes. Antimalarial drugs treat it.",
    "Tuberculosis affects the lungs causing persistent cough and night sweats. Requires 6-month antibiotics.",
    "Dengue fever causes high fever and joint pain. Spread by Aedes mosquitoes.",
    "Heart disease risk factors include high cholesterol and obesity. Managed with medications.",
    "Kidney disease reduces filtering ability. Advanced stages may need dialysis.",
    "Alzheimer's causes memory loss and cognitive decline. No cure exists but medications slow progression.",
]

docs = [Document(page_content=text, metadata={"source": f"doc_{i}"}) for i, text in enumerate(medical_data)]
splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
chunks = splitter.split_documents(docs)

print("Creating embeddings...")
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
vectorstore = Chroma.from_documents(chunks, embeddings)
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0.3)
print("✅ RAG Pipeline ready!")
```
